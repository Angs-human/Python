# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

### What is Matplotlib?

**Matplotlib** is a popular data visualization library in Python that is used to create static, animated, and interactive plots. It provides a flexible and powerful interface for producing a wide variety of visualizations, making it a standard tool for data scientists, engineers, and researchers to explore and present data.

### Why is Matplotlib Used?

Matplotlib is used because it allows users to:
1. **Visualize Data**: It helps in understanding the underlying patterns, trends, and distributions in data.
2. **Present Data**: The plots generated by Matplotlib can be used in presentations, reports, and publications.
3. **Customization**: It provides a high degree of control over plot appearance, including colors, labels, line styles, and more.
4. **Integration**: It integrates well with other libraries like NumPy, Pandas, and SciPy, making it easy to visualize complex data structures.
5. **Compatibility**: It can produce plots in various formats (like PNG, PDF, SVG), suitable for both web and print.

### Five Plots That Can Be Plotted Using the Pyplot Module of Matplotlib:

1. **Line Plot**:
   - Used to display data points connected by straight lines.
   - Useful for visualizing trends over time or other continuous variables.
   ```python
   plt.plot(x, y)
   ```

2. **Bar Plot**:
   - Used to represent categorical data with rectangular bars.
   - The length of the bar represents the magnitude of the variable.
   ```python
   plt.bar(categories, values)
   ```

3. **Histogram**:
   - Used to represent the distribution of numerical data.
   - It divides the data into bins and shows the frequency of each bin.
   ```python
   plt.hist(data, bins=10)
   ```

4. **Scatter Plot**:
   - Used to visualize the relationship between two continuous variables.
   - Each point represents an observation in the dataset.
   ```python
   plt.scatter(x, y)
   ```

5. **Pie Chart**:
   - Used to represent the proportions of different categories within a whole.
   - Each slice of the pie represents a category's contribution to the total.
   ```python
   plt.pie(values, labels=categories)
   ```

These plots are just a few examples of what can be created using Matplotlib's `pyplot` module, which provides a MATLAB-like interface for making plots.

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create a grid of plots within a single figure. It allows you to display multiple plots side by side or in a specified arrangement. This is particularly useful when you want to compare different visualizations or present multiple aspects of the same data in a coherent layout.

````python
import matplotlib.pyplot as plt
import numpy as np

# Data for the lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and a 2x2 grid of subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1, color='blue')
plt.title("Line 1")

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x2, y2, color='green')
plt.title("Line 2")

# Third subplot
plt.subplot(2, 2, 3)
plt.plot(x3, y3, color='red')
plt.title("Line 3")

# Fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x4, y4, color='purple')
plt.title("Line 4")

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
````

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot (or bar chart) is a type of graph that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots can be plotted either vertically (with bars extending upward or downward) or horizontally (with bars extending leftward or rightward).

Why is a Bar Plot Used?
Bar plots are used to:

Compare Categories: Bar plots are ideal for comparing the magnitude of different categories or groups. Each bar represents a category, and its length shows the size or frequency of that category.
Visualize Discrete Data: They are particularly useful for displaying discrete data, where each bar corresponds to a distinct category, like types of products, regions, or age groups.
Show Trends Over Time: When the categories represent time intervals, bar plots can be used to show trends, such as sales over months or years.
Highlight Differences: They make it easy to visually assess differences between categories, which helps in decision-making or identifying patterns.

````python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
````

````python
plt.bar(company,profit)
````

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

What is a Box Plot?
A box plot (also known as a box-and-whisker plot) is a graphical representation of the distribution of a dataset that displays the data's minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It may also show outliers.

Why is a Box Plot Used?
Box plots are used for:

Visualizing the Spread and Skewness: Box plots show the range of the data and how it is distributed across quartiles, providing insights into the skewness and spread of the dataset.
Identifying Outliers: They help in identifying outliers, which are data points that fall outside the typical range.
Comparing Distributions: Multiple box plots can be plotted side by side to compare the distributions of different datasets.
Understanding Central Tendency and Variability: The median, quartiles, and whiskers in a box plot provide a clear summary of the central tendency and variability of the data.

````python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating the box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])

# Adding a title and labels
plt.title("Box Plot Example")
plt.ylabel("Values")

# Show the plot
plt.show()
````
